<b>From above, we can see that there are 3 formats of data types:<b>

object: Object format indicates variables are categorical. Categorical variables in the dataset are city, sex, social_class, primary_business, secondary_business, type_of_house, loan_purpose.

int64: This represents integer variables. Integer variables in the dataset are Id, age, old_dependents, young_dependents, occupants_count, loan_tenure, loan_installments.

float64: This represents variables that have some decimal values involved. Float variables in the dataset are annual-income, monthly_expenses, home_ownership, house_area, sanitary_availability, water_availability, loan_amount.
